#### 7. I have pairgrid using scatter plot and showing according to performance score is correlated with different columns of dataset_2. 

```python
g = sns.PairGrid(dataset_2, vars=['EmpSatisfaction','EngagementSurvey','PerfScoreID','DeptID','ManagerID','Salary'],
                 hue='PerformanceScore', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```
